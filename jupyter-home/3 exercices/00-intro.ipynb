# Intro aux TDs

# Intro
Le but du module est double : lister quelques grandes technologies "big data" (boomer) et sentir leurs forces et faiblesses pour savoir quand les utiliser.

Technologies et implémentations vues ici :
- Data Base (DB) SQL - MySQL
- Calculs distribués - Spark via PySpark
- Index lexical - Vespa
- Index vectoriel - FAISS
- Index hybride - Vespa

Autres technologies évoquées : Cache (REDIS), DB graphe (Neo4J)

# Outils mis à disposition
Un server tournera en permanence en hébergeant les services utiles au TDs : MySQL, Vespa etc ... 
Pour permettre à tout le monde de développer dans les mêmes conditions, un [JupyerHub](https://jupyter.org/hub) permettant de manipuler des [notebooks Jupyer](https://jupyter.org/) Python est mis en place pour tout le monde. Les notebooks communiquent nativement avec tous les services du serveur.

# Ce que le TD n'est pas 
Ce TD n'est pas là pour vous apprendre les couches réseaux, les configuration fines des outils utilisés, Docker, Kubernetes ... Néanmoins certains aspects pourront être brièvement abordés.

# Déroulé des TDs

Chaque TD sera focalisé sur une des technologies ci-dessus. 
Les mêmes données resserviront à chaque TDs - description ci-dessous.
5 use-cases (UC) issus du quotidien professionel seront étudiés à chaque TD - description ci-dessous. Nous essayerons de les traiter grâce à la techno du jour afin d'éprouver la capacité d'une technologie donnée à gérer un type de demande.

## Data proposées

DB `beers` contenant 39 caractéristiques de ~6k bières : propriétés chimiques, type, brasserie, etc ... https://github.com/dcos/demos/tree/master/beer-demo/1.10/database/sql 

## Uses cases abordés

- UC1 : Selection de vues sur les data, exploration data, statistiques : vues, counts, moyennes, médianes, histogrammes, corrélations 
- UC2 : créer un dataset destiné à un apprentissage ML (détails après) 
- UC3 : récupération des docs qui parlent d'un sujet par mot clef
- UC4 : récupérer des attributs textuels des données, les envoyer à un service de vectorisation afin d'utiliser ces vecteurs plus tard
- UC5 : répondre à des questions sur le corpus

# Notation

Basée sur la participation en TD + projet avec soutenance. 

## Projet 
Un petit projet de fin de module sera à réaliser en petits groupes. Il s'agira surtout de recycler des concepts vus au cours des TDs, pas de réinventer la roue. 
**Intérêt principal : combiner 2+ technos évoquées en TD au sein d'une architecture** permettant de répondre élégamment à une problématique data pointue.

**Sujet annoncé plus tard** + possibilité de choisir un autre sujet si tout le groupe est d'accord MAIS attention à la qualité du sujet et au respect de la contrainte "2+ technos combinées"

La qualité de l'architecture du code n'a pas d'importance (un notebook + 1 script suffisent).
## Évaluation continue
- 25% fréquence et pertinence des propositions pendant les TDs
- 75% lisibilité du code + preuve de fonctionnement 

## Évaluation exam
- qualité des arguments qui justifient le recours aux différentes technologies
- mise en avant de la collaboration entre ces technos pour résoudre le problème posé
